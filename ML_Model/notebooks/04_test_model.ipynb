```python
import joblib
import json

# Load production model and tools
model = joblib.load('../models/best_model.pkl')
vectorizer = joblib.load('../models/tfidf_vectorizer.pkl')

with open('../models/label_mapping.json') as f:
    label_mapping = json.load(f)

# Reverse mapping
id_to_category = {v: k for k, v in label_mapping.items()}

print("✅ Model loaded successfully!")
print(f"Model type: {type(model).__name__}")
print(f"\nCategories ({len(id_to_category)}):")
for cat_id, cat_name in sorted(id_to_category.items()):
    print(f"  {cat_id}: {cat_name}")
```

```python
def categorize_text(text):
    """Categorize a single text"""
    # Transform text
    text_vec = vectorizer.transform([text])
    
    # Predict
    prediction = model.predict(text_vec)[0]
    category = id_to_category[prediction]
    
    # Get confidence if available
    if hasattr(model, 'predict_proba'):
        proba = model.predict_proba(text_vec)[0]
        confidence = proba[prediction]
        return category, confidence
    else:
        return category, None

# Test the function
test_text = "How to build a gaming PC step by step"
category, confidence = categorize_text(test_text)

print(f"Test: '{test_text}'")
print(f"→ Category: {category}")
if confidence:
    print(f"→ Confidence: {confidence:.2%}")
```

```python
# Diverse test samples
test_samples = [
    "How to build a gaming PC step by step tutorial for beginners",
    "Best goals and highlights from last night's soccer match",
    "Delicious chocolate cake recipe easy to make at home",
    "Python programming tutorial for absolute beginners",
    "Funny cat videos compilation 2024 try not to laugh",
    "Latest news on climate change and global warming crisis",
    "Learn calculus from scratch mathematics education",
    "Fortnite battle royale gameplay season 5",
    "How to lose weight fast workout routine",
    "Taylor Swift new music video behind the scenes"
]

print("="*70)
print("TESTING MODEL ON SAMPLE TEXTS")
print("="*70)

for i, text in enumerate(test_samples, 1):
    category, confidence = categorize_text(text)
    print(f"\n{i}. Text: {text}")
    print(f"   → Category: {category}")
    if confidence:
        print(f"   → Confidence: {confidence:.2%}")

print("\n" + "="*70)
print("✅ Model testing complete!")
print("="*70)
```
